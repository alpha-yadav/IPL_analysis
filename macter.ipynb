### 2. POINT TABLE OF EVERY SEASON

```python
import datetime as dt
def points_table(matches):
    teams = pd.concat([matches["team1"], matches["team2"]]).unique()
    table = pd.DataFrame(index=teams)
    table["Played"] = 0
    table["Won"] = 0
    table["Lost"] = 0
    table["No Result"] = 0
    table["Points"] = 0

    for _, row in matches.iterrows():

        team1 = row["team1"]
        team2 = row["team2"]
        winner = row["winner"]
        result = row["result"]

        table.loc[team1, "Played"] += 1
        table.loc[team2, "Played"] += 1

        if pd.isna(winner):
            table.loc[team1, "No Result"] += 1
            table.loc[team2, "No Result"] += 1

            table.loc[team1, "Points"] += 1
            table.loc[team2, "Points"] += 1

        else:
            table.loc[winner, "Won"] += 1
            table.loc[winner, "Points"] += 2

            loser = team2 if winner == team1 else team1
            table.loc[loser, "Lost"] += 1
    table = table.sort_values(["Points", "Won"], ascending=False)

    return table.reset_index().rename(columns={"index": "Team"})
t2='POINT TABLE OF EVERY SEASON'
title(t2)
for i in df_mat["season"].unique():
    print("-"*81)
    print("\t\t\t\t",i)
    print("-"*81)
    points = points_table(df_mat[df_mat["season"] == i])
    print(points)
    
```
